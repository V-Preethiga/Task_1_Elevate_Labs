```python
# Load the dataset
df = pd.read_csv("/content/Sales_with_NaNs_v1.3.csv")  # replace with your actual file name

# Preview first 5 rows
df.head()
```

```python
# Number of rows and columns
print(df.shape)

# Column names and data types
print(df.info())

# Count missing values per column
print(df.isnull().sum())
```

```python
# Fill numeric columns with mean
numeric_cols = ['Sales_Before', 'Sales_After', 'Customer_Satisfaction_Before', 'Customer_Satisfaction_After']
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())

# Fill categorical columns with 'Unknown'
categorical_cols = ['Group', 'Customer_Segment', 'Purchase_Made']
for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Remove extra spaces and lowercase for uniformity
for col in categorical_cols:
    df[col] = df[col].str.strip().str.lower()
```

```python
df['Sales_Change'] = df['Sales_After'] - df['Sales_Before']
```

```python
df['Satisfaction_Change'] = df['Customer_Satisfaction_After'] - df['Customer_Satisfaction_Before']
```

```python
# Example: yes=1, no=0
df['Purchase_Made_Numeric'] = df['Purchase_Made'].map({'yes':1, 'no':0, 'unknown':0})
```

```python
df.to_csv("sales_cleaned.csv", index=False)
```
